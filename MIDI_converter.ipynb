```python
from midiutil import MIDIFile
from dic import composers
import pandas as pd
import os
```

```python
def serie_a_midi(serie, archivo_salida, duracion_nota=1, volumen=100, tempo=120):
    # Crear un nuevo archivo MIDI
    midi = MIDIFile(1)

    # Configurar algunos parmetros del archivo MIDI
    midi.addTempo(0, 0, tempo)

    # Convertir las notas a pitch MIDI y agregar eventos al archivo MIDI
    tiempo = 0
    for pitch in serie:
        midi.addNote(0, 0, int(pitch), tiempo, duracion_nota, volumen)
        tiempo += duracion_nota

    # Guardar el archivo MIDI
    with open(archivo_salida, 'wb') as file:
        midi.writeFile(file)

    print(f"Archivo MIDI guardado como '{archivo_salida}'")


for composer in composers.keys():
    for pieza in composers[composer]:
        
        serie_a_midi(composers[composer][pieza], "MIDI/" + str(composer) + '/' + str(pieza) + '.mid', duracion_nota=1, volumen=100, tempo=120)
```

```python
import mido

def midi_a_txt(archivo_midi, archivo_txt):
    # Abrir el archivo MIDI
    mid = mido.MidiFile(archivo_midi)

    # Lista para almacenar los nmeros de las notas
    notas = []

    # Iterar sobre cada pista en el archivo MIDI
    for track in mid.tracks:
        # Iterar sobre cada mensaje en la pista
        for msg in track:
            # Verificar si el mensaje es de tipo "note_on" (nota activada)
            if msg.type == 'note_on':
                # Agregar el nmero de la nota a la lista
                notas.append(msg.note)

    # Escribir los nmeros de las notas en el archivo de texto
    with open(archivo_txt, 'w') as file:
        for nota in notas:
            file.write(f'{nota}\n')


midi_a_txt("Mozart_Sonata.mid", "Mozart_Sonata.txt")
```
